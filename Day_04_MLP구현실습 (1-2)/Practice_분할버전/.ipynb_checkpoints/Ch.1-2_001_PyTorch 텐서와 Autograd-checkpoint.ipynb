# PyTorch н…җм„ң (Tensor)
> н…җм„ңлҠ” <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">лӢӨм°Ёмӣҗ л°°м—ҙ</mark>лЎң, NumPyмқҳ ndarrayмҷҖ мң мӮ¬н•ҳм§Җл§Ң <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">GPU м—°мӮ°кіј мһҗлҸҷ лҜёл¶„</mark>мқ„ м§Җмӣҗн•©лӢҲлӢӨ.

><mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">NumPy</mark>лҠ” CPU кё°л°ҳ мҲҳм№ҳ м—°мӮ°м—җ нҺёлҰ¬н•ҳм§Җл§Ң, л”Ҙлҹ¬лӢқм—җлҠ” л‘җ к°Җм§Җ к·јліём Ғ н•ңкі„к°Җ мһҲмҠөлӢҲлӢӨ.</br>
> мІ«м§ё, GPU м—°мӮ°мқҙ л¶Ҳк°ҖлҠҘн•ҳм—¬ мҲҳл°ұл§Ң нҢҢлқјлҜён„°мқҳ н–үл ¬ кіұмқ„ CPUл§ҢмңјлЎң мІҳлҰ¬н•ҳл©ҙ мҲҳ мӢңк°„~мҲҳ мқјмқҙ кұёлҰ¬лҠ” л°ҳл©ҙ, <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">GPU лі‘л ¬ м—°мӮ°мқҖ мҲҳ л¶„~мҲҳмӢӯ л¶„</mark>м—җ мІҳлҰ¬н•©лӢҲлӢӨ.</br>
> л‘ҳм§ё, мһҗлҸҷ лҜёл¶„(Autograd)мқҙ м—Ҷм–ҙ <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">м—ӯм „нҢҢ(Backpropagation)</mark>м—җ н•„мҡ”н•ң мҲҳл°ұл§Ң нҺёлҜёл¶„мқ„ мҲҳлҸҷмңјлЎң кө¬н•ҙм•ј н•©лӢҲлӢӨ.</br></br>
> PyTorchлҠ” `.to("cuda")` н•ң мӨ„лЎң GPU м „нҷҳмқҙ к°ҖлҠҘн•ҳкі , `backward()` нҳём¶ң н•ң лІҲмңјлЎң лӘЁл“  к·ёлһҳл””м–ёнҠёлҘј мһҗлҸҷ кі„мӮ°н•©лӢҲлӢӨ. мҰү, PyTorch = NumPyмқҳ нҺёмқҳм„ұ + GPU к°ҖмҶҚ + мһҗлҸҷ лҜёл¶„мһ…лӢҲлӢӨ.

> мӢӨм ң мӢ кІҪл§қм—җм„ң нҢҢлқјлҜён„° мҲҳлҠ” мҲҳл°ұл§Ң~мҲҳмӢӯм–ө к°ңм—җ лӢ¬н•ҳлҜҖлЎң, к°Ғ нҢҢлқјлҜён„°м—җ лҢҖн•ҙ мҶҗмңјлЎң лҜёл¶„мӢқмқ„ мң лҸ„н•ҳкі  мҪ”л“ңлЎң кө¬нҳ„н•ҳлҠ” кІғмқҖ нҳ„мӢӨм ҒмңјлЎң л¶Ҳк°ҖлҠҘн•©лӢҲлӢӨ.</br>
> л Ҳмқҙм–ҙ м¶”к°Җ мӢң лҜёл¶„мӢқ мһ¬мң лҸ„, кі„мӮ° мӢӨмҲҳлЎң мқён•ң л””лІ„к№… лӮңмқҙлҸ„, лӘЁлҚё кө¬мЎ° ліҖкІҪ мӢң лҜёл¶„ мҪ”л“ң м „мІҙ мҲҳм • л“ұмқҳ л¬ём ңк°Җ мһҲмңјл©°, <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">AutogradлҠ” мқҙ лӘЁл“  кіјм •мқ„ м—°мӮ° к·ёлһҳн”„ м¶”м ҒмңјлЎң мһҗлҸҷнҷ”</mark>н•ҳм—¬ лӘЁлҚё м„Өкі„м—җл§Ң м§‘мӨ‘н•  мҲҳ мһҲкІҢ н•©лӢҲлӢӨ.
