## 페르소나 기반 생성 (Persona-based Generation)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">가상의 인물 프로필을 설정</mark>하여 다양하고 현실적인 데이터를 생성합니다.

````python
# TODO 1: 30대 직장인 여성 '김미래' 페르소나를 설정하고, 무선 블루투스 이어폰에 대한 별점 4점 리뷰를 요청하는 프롬프트를 작성하여 LLM을 호출해봅시다.

prompt = """당신은 30대 직장인 여성 '김미래'입니다.
온라인 쇼핑을 자주 하며 가격 대비 품질을 중시합니다.

다음 상품에 대한 리뷰를 작성해주세요:
상품: 무선 블루투스 이어폰
별점: 4점

리뷰:"""

response = llm.invoke(prompt)
print(response.content)
````

💡페르소나 다양성
> 연령, 직업, 성격, 구매 패턴 등을 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">다양하게 조합</mark>하면</br>
> 편향되지 않은 균형 잡힌 합성 데이터를 생성할 수 있습니다.

## JSON 구조화 출력
> LLM의 출력을 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">프로그래밍적으로 파싱 가능한 JSON 형식</mark>으로 받습니다.

````python
# TODO 2: JSON 형식으로 고객 리뷰 3개를 생성하도록 요청하는 프롬프트를 작성해봅시다.

prompt = """다음 형식의 JSON으로 고객 리뷰 데이터 3개를 생성하세요:

{
    "reviews": [
        {
            "customer_name": "이름",
            "rating": 1-5,
            "category": "긍정/부정/중립",
            "text": "리뷰 내용"
        }
    ]
}"""
````

````python
# TODO 3: LLM을 호출한 뒤 응답을 파싱하여 출력해봅시다.

response = llm.invoke(prompt)

import json
result = json.loads(response.content)
for r in result["reviews"]:
    print(f"[{r['category']}] {r['customer_name']} ({r['rating']}점): {r['text'][:30]}...")
````

💡JSON 파싱 안정성
> LLM이 항상 유효한 JSON을 생성하지는 않습니다.</br>
> `try-except`로 파싱 오류를 처리하고, 필요시 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">재생성을 요청</mark>하세요.

## Structured Output (response_format)
> 프롬프트로 JSON 형식을 "요청"하는 것과 달리, <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">response_format 파라미터로 JSON Schema를 강제</mark>하면 API가 항상 유효한 JSON을 반환합니다.

앞서 TODO 2에서는 프롬프트에 JSON 형식을 명시하여 LLM에게 구조화된 출력을 "요청"했습니다. 하지만 이 방식은 LLM이 지시를 따르지 않거나 불완전한 JSON을 생성할 위험이 있습니다.

OpenAI 호환 API의 `response_format` 파라미터를 사용하면 다음과 같은 이점이 있습니다:

<table style="width:100%">
  <thead>
    <tr>
      <th style="text-align:center">비교 항목</th>
      <th style="text-align:center">프롬프트 기반 (TODO 2)</th>
      <th style="text-align:center">response_format (TODO 3)</th>
    </tr>
  </thead>
  <tbody>
    <tr><td style="text-align:center">JSON 보장</td><td style="text-align:center">보장 안 됨 (LLM 재량)</td><td style="text-align:center">API 레벨에서 강제</td></tr>
    <tr><td style="text-align:center">필드 누락</td><td style="text-align:center">발생 가능</td><td style="text-align:center">required 필드 보장</td></tr>
    <tr><td style="text-align:center">파싱 에러</td><td style="text-align:center">try-except 필요</td><td style="text-align:center">거의 발생 안 함</td></tr>
    <tr><td style="text-align:center">스키마 검증</td><td style="text-align:center">별도 검증 코드 필요</td><td style="text-align:center">자동 검증</td></tr>
  </tbody>
</table>

`response_format`에 전달하는 JSON Schema는 다음 구조를 따릅니다:

```python
response_format = {
    "type": "json_schema",
    "json_schema": {
        "name": "스키마_이름",       # 스키마 식별자
        "strict": True,              # 스키마 엄격 준수
        "schema": {
            "type": "object",
            "properties": {          # 출력할 필드 정의
                "필드명": {
                    "type": "string / integer / array / ...",
                    "description": "필드 설명"
                }
            },
            "required": ["필드명"],  # 필수 필드 목록
            "additionalProperties": False  # 정의되지 않은 필드 차단
        }
    }
}
```

> `strict: True`로 설정하면 LLM이 스키마에 정의된 필드만 포함한 JSON을 생성하도록 강제합니다.
> `additionalProperties: False`는 정의되지 않은 추가 필드가 출력에 포함되는 것을 방지합니다.

````python
# TODO 4: 리뷰 데이터를 위한 JSON Schema를 정의해봅시다.

review_response_format = {
    "type": "json_schema",
    "json_schema": {
        "name": "customer_reviews",
        "strict": True,
        "schema": {
            "type": "object",
            "properties": {
                "reviews": {
                    "type": "array",
                    "description": "고객 리뷰 목록",
                    "items": {
                        "type": "object",
                        "properties": {
                            "customer_name": {
                                "type": "string",
                                "description": "고객 이름"
                            },
                            "rating": {
                                "type": "integer",
                                "description": "별점 (1-5)"
                            },
                            "category": {
                                "type": "string",
                                "description": "감성 분류 (긍정/부정/중립)"
                            },
                            "text": {
                                "type": "string",
                                "description": "리뷰 본문"
                            }
                        },
                        "required": ["customer_name", "rating", "category", "text"],
                        "additionalProperties": False
                    }
                }
            },
            "required": ["reviews"],
            "additionalProperties": False
        }
    }
}
````

````python
# TODO 5: 위 스키마를 response_format으로 전달하여 LLM을 호출해봅시다.

prompt = "무선 블루투스 이어폰에 대한 고객 리뷰 3개를 생성하세요. 긍정, 부정, 중립 각 1개씩 포함하세요."

response = llm.invoke(
    prompt,
    response_format=review_response_format
)
````

````python
# TODO 6: 응답을 파싱하여 출력해봅시다.

result = json.loads(response.content)
for r in result["reviews"]:
    print(f"[{r['category']}] {r['customer_name']} ({r['rating']}점): {r['text'][:50]}...")
````

💡response_format 활용 팁
> `response_format`은 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">합성 데이터 파이프라인에서 특히 강력</mark>합니다.</br>
> 수백~수천 건의 데이터를 반복 생성할 때, 프롬프트 기반 JSON은 일부 응답에서 파싱 에러가 발생할 수 있지만</br>
> `response_format`을 사용하면 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">모든 응답이 동일한 스키마를 따르므로 후처리 코드가 단순</mark>해집니다.</br>
> 단, 모든 LLM API가 이 기능을 지원하는 것은 아니므로 사용 전 API 문서를 확인하세요.

````python
# TODO 7: 페르소나 목록과 상품 목록을 이중 반복문으로 순회하며, 각 조합마다 LLM을 호출하여 리뷰를 생성하고 결과를 JSON 파일로 저장해봅시다.

personas = ["20대 대학생", "40대 주부", "60대 은퇴자"]
products = ["노트북", "운동화"]

results = []
for persona in personas:
    for product in products:
        prompt = f"{persona}의 관점에서 {product} 리뷰를 작성하세요."
        review = llm.invoke(prompt).content
        results.append({"persona": persona, "product": product, "review": review})

print(f"생성된 리뷰 수: {len(results)}개")
for r in results[:2]:
    print(f"[{r['persona']}] {r['product']}: {r['review'][:40]}...")

# 결과를 JSON 파일로 저장
with open("synthetic_reviews.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=2)
print(f"\nsynthetic_reviews.json 저장 완료 ({len(results)}건)")
````
