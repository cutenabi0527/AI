# 환경변수 관리 (Environment Variables)
> API 키와 같은 민감한 정보를 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">코드에 직접 노출하지 않고 안전하게 관리</mark>하는 방법입니다.

> LLM API를 사용하려면 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">API 키(비밀번호와 동일)</mark>가 필요합니다.</br></br>
> 이 키를 코드에 직접 작성하면 GitHub 등 버전 관리 시스템에 올라가는 순간 누구나 볼 수 있게 되며, 실제로 노출된 API 키는 수분 내에 자동화된 봇에 의해 탈취되어 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">막대한 비용 피해</mark>로 이어질 수 있습니다.</br></br>
> 환경변수 방식을 사용하면 코드와 비밀 정보를 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">완전히 분리</mark>할 수 있습니다.</br></br>
> `.env` 파일(숨김 파일)은 `.gitignore`에 추가해 버전 관리에서 제외하고, 팀원마다 각자의 키를 각자의 환경에 설정하므로 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">공유 없이 협업</mark>이 가능합니다.</br></br>
> 배포 환경(서버, 클라우드)에서도 동일한 코드로 다른 키를 사용할 수 있어, 운영체제가 프로세스에 전달하는 키-값 쌍인 환경변수는 API 키 관리의 표준적인 방법입니다.

</br>

## .env 파일
> 프로젝트 루트에 `.env` 파일을 생성하여 환경변수를 저장합니다.

💡.env 보안
> `.env` 파일은 반드시 `.gitignore`에 추가하여 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">버전 관리에서 제외</mark>해야 합니다.</br>
> API 키가 GitHub에 노출되면 즉시 탈취될 수 있습니다.

</br>

## python-dotenv
> `.env` 파일의 환경변수를 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">Python에서 자동으로 로드</mark>합니다.

```python
# TODO 1: .env 파일을 로드하고, 환경변수에서 API 키를 읽어 변수에 저장한 뒤 결과를 출력해봅시다.

from dotenv import load_dotenv
import os

# .env 파일 로드
load_dotenv()

# 환경변수 읽기
api_key = os.getenv("UPSTAGE_API_KEY")
print(f"API Key 로드: {'성공' if api_key else '실패'}")
print(f"키 앞 5자리: {api_key[:5]}..." if api_key else "키 없음")
```

</br>

## os.environ으로 직접 설정

```python
# TODO 2: 환경변수에 API 키를 직접 설정하고, 읽어서 변수에 저장한 뒤 앞 10자리를 출력해봅시다.

import os

# 직접 설정 (개발/테스트 용도)
os.environ["UPSTAGE_API_KEY"] = "your-api-key"

# 환경변수 읽기
key = os.environ.get("UPSTAGE_API_KEY", "default-value")
print(f"설정된 키: {key[:10]}...")
```
