# NumPy 배열 (ndarray)
> NumPy는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">고성능 다차원 배열(ndarray)</mark>을 제공하는 수치 연산 라이브러리입니다.</br>
> Python 리스트보다 메모리 효율적이고 연산 속도가 빠릅니다.

>Python 리스트로 백만 개의 숫자를 2배로 만들려면 `for` 반복문이 필요합니다. 느립니다.</br>
>NumPy는 같은 작업을 `arr * 2` 한 줄로, <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">내부적으로 C 언어 수준의 속도</mark>로 처리합니다.</br>
>실제 벤치마크에서 NumPy는 Python 리스트보다 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">수십 배에서 100배 이상 빠릅니다</mark>.</br>
>메모리 측면에서도 Python 리스트는 각 숫자를 객체로 저장하지만, NumPy 배열은 연속된 메모리 블록에 순수 숫자만 저장하여 훨씬 적은 메모리를 사용합니다.

><mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">벡터화</mark>란 반복문 없이 배열 전체에 한 번에 연산을 적용하는 방식입니다.</br>
머신러닝에서 가중치 업데이트, 데이터 정규화, 행렬 곱셈 등의 연산은 수백만 번 반복되는데, Python `for` 루프로는 이런 연산이 실용적인 시간 안에 끝나지 않습니다.</br>
>NumPy의 벡터화 연산은 Pandas, TensorFlow, scikit-learn 등 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">데이터 과학 생태계 전체의 기반</mark>이며, NumPy를 이해하면 이 도구들의 동작 방식을 훨씬 깊이 이해할 수 있습니다.

## 배열 연산 (Vectorized Operations)
> NumPy는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">반복문 없이 배열 전체에 연산</mark>을 적용합니다.

```python
# TODO 8: array_a에 1x3 크기의 1, 2, 3 행렬을, array_b에 1x3 크기의 4, 5, 6 행렬을 저장하고 합, 곱, 제곱, 내적을 출력해봅시다.
arr_a = np.array([1,2,3])
arr_b = np.array([4,5,6])

arr_a + arr_b
arr_a * arr_b
arr_a ** 2
np.dot(arr_a, arr_b)

```

```python
# TODO 9: 1x5 크기의 3, 1, 4, 1, 5 데이터를 array에 저장하고, 최댓값, 최솟값, 최솟값 인덱스, 정렬 결과를 출력해봅시다.

array = np.array([3, 1, 4, 1, 5])
print(f"최댓값: {np.max(array)}")
print(f"최솟값: {np.min(array)}")
print(f"최솟값 인덱스: {np.argmin(array)}")
print(f"정렬: {np.sort(array)}")
```

## 브로드캐스팅 (Broadcasting)
> 크기가 다른 배열 간 연산 시 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">작은 배열을 자동으로 확장</mark>하는 규칙입니다.

```python
# TODO 10: matrix에 2x3 크기의 행렬에 1부터 6까지 행렬을 저장하고, vector에 1x3 크기의 행렬에 10, 20, 30을 저장한 후 matrix + vector 브로드캐스팅 결과를 출력해봅시다.
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

vector = np.array([10, 20, 30])

result = matrix + vector
print(result)
```

💡브로드캐스팅 규칙
> 뒤쪽 차원부터 비교하여 크기가 같거나 하나가 1이면 브로드캐스팅이 적용됩니다.

## 집계 함수 (Aggregation)

```python
# TODO 11: array에 2x3 크기의 행렬에 1부터 6까지 저장하고, 전체 합, 열별 합 axis =0 , 행별 평균 axis=1, 최댓값 인덱스를 출력해봅시다.
array = np.array([[1, 2, 3],
                  [4, 5, 6]])

print(f"전체 합: {np.sum(array)}")
print(f"열별 합: {np.sum(array, axis=0)}")
print(f"행별 평균: {np.mean(array, axis=1)}")
print(f"최댓값 인덱스: {np.argmax(array)}")
```

💡axis 파라미터
> `axis=0`은 행 방향(열별 집계), `axis=1`은 열 방향(행별 집계)입니다.

## 수학 함수 및 비교 (Math Functions & Comparison)
> NumPy는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">원소별 수학 함수</mark>와 배열 비교 함수를 제공합니다.

<table style="width:100%">
  <thead>
    <tr>
      <th style="text-align:center">함수</th>
      <th>설명</th>
    </tr>
  </thead>
  <tbody>
    <tr><td style="text-align:center"><code>np.sin(arr)</code> / <code>np.cos(arr)</code></td><td>원소별 삼각함수 적용</td></tr>
    <tr><td style="text-align:center"><code>np.exp(arr)</code></td><td>원소별 지수함수 (e^x) 적용</td></tr>
    <tr><td style="text-align:center"><code>np.log(arr)</code></td><td>원소별 자연로그 적용</td></tr>
    <tr><td style="text-align:center"><code>np.array_equal(a, b)</code></td><td>두 배열이 완전히 같은지 비교</td></tr>
    <tr><td style="text-align:center"><code>np.allclose(a, b)</code></td><td>두 배열이 허용 오차 내에서 같은지 비교</td></tr>
  </tbody>

```python
# TODO 12:
#  - 1x3 크기의 행렬에 0, π/2, π을 angles에 저장하고 sin, cos 출력
#  - 1x3 크기의 행렬에 1, 2, 3을 vals에 저장하고 지수, 로그를 출력
#  - a, b 비교, a, c 비교, 허용 오차 내 a, c 비교
#  를 출력해봅시다.
angles = np.array([0, np.pi/2, np.pi])
print(f"sin: {np.sin(angles)}")
print(f"cos: {np.cos(angles)}")

vals = np.array([1, 2, 3])
print(f"지수: {np.exp(vals)}")
print(f"로그: {np.log(vals)}")

a = np.array([1, 2, 3])
b = np.array([1, 2, 3])
c = np.array([1.0, 2.0, 3.0000001])
print(np.array_equal(a, b))
print(np.array_equal(a, c))
print(np.allclose(a, c))
```

💡array_equal vs allclose
> `np.array_equal()`은 값이 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">정확히 일치</mark>해야 True를 반환합니다.</br>
> `np.allclose()`는 부동소수점 오차를 허용하므로 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">수치 계산 결과 비교</mark>에 적합합니다.
