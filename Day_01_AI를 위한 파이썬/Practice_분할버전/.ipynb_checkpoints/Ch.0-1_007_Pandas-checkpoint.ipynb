# Pandas 기본 구조
> Pandas는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">표 형태의 데이터를 다루는</mark> Python 라이브러리입니다.

>NumPy는 숫자 연산에 최적화되어 있지만, 현실의 데이터는 훨씬 복잡합니다.</br>
이름, 나이, 도시처럼 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">서로 다른 타입이 섞인 데이터</mark>를 NumPy 배열 하나로 표현하기 어렵고, 열 이름 개념이 없어 `arr[:, 2]`처럼 숫자 인덱스로만 접근해야 합니다.</br>
>결측치(`NaN`) 처리, 날짜 데이터, CSV/Excel 파일 입출력 등의 기능도 NumPy에는 없습니다.

>Pandas는 이런 문제들을 해결합니다. <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">열 이름으로 접근</mark>하여 `df["age"]`처럼 의미 있는 이름으로 데이터를 선택할 수 있고, 숫자, 문자열, 날짜를 하나의 테이블에 담는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">혼합 타입</mark>을 지원합니다.</br>
>`NaN`을 인식하고 `dropna()`, `fillna()`로 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">결측치를 자동 처리</mark>하며, CSV, Excel, JSON, SQL 등을 한두 줄로 읽고 쓸 수 있습니다.</br>
>SQL의 `GROUP BY`처럼 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">강력한 그룹 연산</mark>으로 데이터를 집계하고 분석할 수 있어, Pandas는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">데이터 분석의 엑셀이라고 불릴 만큼</mark> 데이터 탐색과 전처리의 표준 도구입니다.

```python
# TODO 1: 10, 20, 30 값에 a, b, c를 인덱스로 지정한 시리즈를 series에 저장하고 출력해봅시다.
import pandas as pd
# 1. 데이터와 인덱스를 지정하여 시리즈 생성
# 10, 20, 30 데이터에 'a', 'b', 'c' 인덱스를 부여합니다.
series = pd.Series([10, 20, 30], index=['a', 'b', 'c'])

# 2. 결과 출력
print("--- 생성된 시리즈 ---")
print(series)

# 3. 시리즈의 정보 확인 (선택 사항)
print(f"\n인덱스 확인: {series.index}")
print(f"데이터 값 확인: {series.values}")
```

## 결측치 처리

<table style="width:100%">
  <thead>
    <tr>
      <th style="text-align:center">메서드</th>
      <th>설명</th>
    </tr>
  </thead>
  <tbody>
    <tr><td style="text-align:center"><code>df.isnull().sum()</code></td><td>열별 결측치 개수</td></tr>
    <tr><td style="text-align:center"><code>df.dropna()</code></td><td>결측치 행 제거</td></tr>
    <tr><td style="text-align:center"><code>df.fillna(value)</code></td><td>결측치 값으로 대체</td></tr>
  </tbody>
</table>

```python
# TODO 8: name에 None, score에 결측값이 포함된 데이터프레임 df_missing을 생성하고, 
# 열별 결측치 수를 확인한 후, name은 "Unknown", score는 0으로 결측치를 채운 결과를 출력해봅시다.
import numpy as np

# 1. 결측치가 포함된 데이터프레임 생성
df_missing = pd.DataFrame({
    "name": ["Alice", "Bob", None, "David", "Eve"],
    "score": [95, np.nan, 88, np.nan, 70]
})

# 2. 열별 결측치 수 확인 
missing_counts = df_missing.isnull().sum() # .isna()도 가능

# 3. 결측치 채우기 (fillna)
# name은 "Unknown", score는 0으로 채움
df_filled = df_missing.fillna({
    "name": "Unknown",
    "score": 0
})

# 4. 결측치가 포함된 행 삭제 (dropna)
# 결측치가 하나라도 포함된 행은 모두 제거합니다.
# 실제로 제거한 것은 아니라 복제본 생성, 처리 후 반환한 것
df_dropped = df_missing.dropna()

# 결과 출력
print("--- [1] 원본 데이터프레임 ---")
print(df_missing)

print("\n--- [2] 열별 결측치 개수 ---")
print(missing_counts)

print("\n--- [3] 결측치 채우기 (fillna) 결과 ---")
print(df_filled)

print("\n--- [4] 결측치 삭제 (dropna) 결과 ---")
print(df_dropped)
```

## GroupBy (그룹 연산)
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">특정 열을 기준으로 데이터를 그룹화</mark>하고 집계 연산을 수행합니다.

```python
# TODO 9: city와 age 열로 구성된 df_city를 생성하고, 도시별로 그룹화하여 평균 나이와 개수를 출력해봅시다.
# 1. city와 age 열로 구성된 데이터프레임 생성
df_city = pd.DataFrame({
    "city": ["서울", "부산", "서울", "대전", "부산", "서울"],
    "age": [25, 30, 35, 28, 22, 40]
})

# 2. 도시별로 그룹화하여 평균 나이와 개수 계산
# agg() 함수를 사용하면 여러 개의 집계 함수를 한 번에 적용할 수 있습니다.
city_analysis = df_city.groupby("city")["age"].agg(["mean", "count"])

# 컬럼 이름을 보기 좋게 변경 (선택 사항)
city_analysis.columns = ["평균 나이", "인원수"]

# 결과 출력
print("--- [1] 원본 데이터 ---")
print(df_city)

print("\n--- [2] 도시별 그룹화 분석 결과 ---")
print(city_analysis)
```

</br>

## Pivot Table
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">행/열 기준을 지정하여 데이터를 재배열</mark>하는 테이블입니다.

```python
# TODO 10: city, grade, score 열로 구성된 df_scores를 생성하고, 도시를 행, 등급을 열로 하여 점수의 평균을 피벗 테이블로 만들어 출력해봅시다.
# 1. 데이터프레임 생성
df_scores = pd.DataFrame({
    "city": ["서울", "서울", "부산", "부산", "서울", "대전", "대전", "부산"],
    "grade": ["A", "B", "A", "A", "B", "B", "A", "B"],
    "score": [95, 80, 88, 92, 85, 78, 90, 82]
})

# 2. 피벗 테이블 생성
# index: 행 방향으로 세울 기준 (도시)
# columns: 열 방향으로 세울 기준 (등급)
# values: 집계할 대상 (점수)
# aggfunc: 집계 방식 (평균)
pivot_result = df_scores.pivot_table(index="city", columns="grade", values="score", aggfunc="mean")

# 결과 출력
print("--- [1] 원본 데이터 ---")
print(df_scores)

print("\n--- [2] 도시별/등급별 점수 평균 피벗 테이블 ---")
print(pivot_result)
```

</br>

## Melt (언피벗)
> 넓은 형태(wide)의 데이터를 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">긴 형태(long)</mark>로 변환합니다.

```python
# TODO 11: name, math, eng 열로 구성된 wide_df를 생성하고, 
# name을 기준으로 math와 eng를 과목(subject)과 점수(score)로 긴 형태로 변환하여 출력해봅시다.
# 1. 가로로 긴 데이터프레임(Wide format) 생성
wide_df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "math": [90, 80, 70],
    "eng": [85, 95, 75]
})

# 2. 긴 형태(Long format)로 변환
# id_vars: 고정할 열 (기준점)
# value_vars: 행으로 녹여낼(melt) 열들 (지정하지 않으면 id_vars를 제외한 모든 열 대상)
# var_name: 변수들이 들어갈 열 이름 (과목)
# value_name: 값들이 들어갈 열 이름 (점수)
long_df = wide_df.melt(id_vars="name", 
                       var_name="subject", 
                       value_name="score")

# 결과 출력
print("--- [1] 원본 데이터 (Wide Format) ---")
print(wide_df)

print("\n" + "="*40 + "\n")

print("--- [2] 변환된 데이터 (Long Format) ---")
print(long_df)
```

💡pivot vs melt
> `pivot_table`은 Long → Wide 변환, `melt`는 Wide → Long 변환입니다.</br>
> 두 연산은 서로 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">역변환</mark> 관계입니다.

```python
# TODO 12: df를 age 기준으로 내림차순 정렬하여 출력해봅시다.
# 기초 데이터 준비
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "age": [25, 30, 35, 28, 22],
    "city": ["서울", "부산", "대전", "광주", "인천"]
}
df_base = pd.DataFrame(data)

# ascending=False를 설정하면 큰 값부터 정렬됩니다.
df_sorted = df_base.sort_values(by="age", ascending=False)
print("--- 나이 기준 내림차순 정렬 ---")
print(df_sorted)
print("\n")
```

```python
# TODO 13: df_base에서 name을 인덱스로 설정하고, 다시 reset_index()로 초기화하여 출력해봅시다.
# set_index: 특정 열을 인덱스로 만듭니다.
df_indexed = df_base.set_index("name")
print("--- 'name'을 인덱스로 설정 ---")
print(df_indexed)

# reset_index: 인덱스를 다시 일반 열로 되돌리고 숫자 인덱스를 생성합니다.
df_reset = df_indexed.reset_index()
print("\n--- 인덱스 초기화(reset_index) ---")
print(df_reset)
print("\n")
```

```python
# TODO 14: df_base의 열 이름을 name→성명, age→나이, city→도시로 변경하고, age 열을 float으로 변환하여 출력해봅시다.
# rename: 딕셔너리 형태로 이름을 바꿉니다.
# astype: 열의 데이터 타입을 강제로 변환합니다.
df_renamed = df_base.rename(columns={"name": "성명", "age": "나이", "city": "도시"})
df_renamed["나이"] = df_renamed["나이"].astype(float)

print("--- 열 이름 변경 및 타입 변환(float) ---")
print(df_renamed)
print(f"\n변경 후 '나이' 열 타입: {df_renamed['나이'].dtype}")
```

💡 sort_values vs sort_index
> `sort_values(by=)`는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">열의 값</mark> 기준, `sort_index()`는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">인덱스</mark> 기준으로 정렬합니다.

      <th>설명</th>
    </tr>

```python
# TODO 15: df_base의 age 열에 lambda를 사용하여 2배 값을 age_doubled 열로 추가하고 출력해봅시다.
# apply() 함수와 lambda를 결합하여 기존 열의 값을 변형해 새 열을 만듭니다.
df_base["age_doubled"] = df_base["age"].apply(lambda x: x * 2)

print("--- age_doubled 열 추가 결과 ---")
print(df_base)
print("\n")
```

```python
# TODO 16: name과 score 열로 구성된 두 DataFrame을 pd.concat()으로 연결하고 인덱스를 초기화하여 출력해봅시다.
# 연결할 두 데이터프레임 생성
df1 = pd.DataFrame({"name": ["David", "Eve"], "score": [85, 90]})
df2 = pd.DataFrame({"name": ["Frank", "Grace"], "score": [75, 80]})

# pd.concat: 리스트 형태로 합칠 객체들을 전달합니다.
# ignore_index=True: 기존 인덱스를 무시하고 0부터 새로 번호를 매깁니다.
df_combined = pd.concat([df1, df2], ignore_index=True)

print("--- 데이터프레임 연결(concat) 결과 ---")
print(df_combined)
```

💡 apply 활용
> `apply()`는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">lambda 또는 사용자 정의 함수</mark>를 열/행 단위로 적용할 수 있습니다.

</br>

## Merge (병합)
> 두 DataFrame을 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">공통 열(키)을 기준으로 결합</mark>합니다.

```python
# TODO 17: id와 name 열의 df_name, id와 score 열의 df_score를 생성하고, id를 기준으로 내부 조인하여 병합한 결과를 출력해봅시다.
# 1. 기초 데이터프레임 생성
# 공통된 열인 'id'를 포함하여 각각의 정보를 담은 표를 만듭니다.
df_name = pd.DataFrame({
    "id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Charlie", "David"]
})

df_score = pd.DataFrame({
    "id": [1, 2, 3, 5], # 4번은 점수가 없고, 5번은 이름 데이터에 없음
    "score": [95, 80, 88, 70]
})

# 2. 'id'를 기준으로 내부 조인(Inner Join) 수행
# how='inner': 두 데이터프레임 모두에 존재하는 'id' 값만 남깁니다.
df_merged = pd.merge(df_name, df_score, on="id", how="inner")

# 결과 출력
print("--- [1] 이름 데이터 (df_name) ---")
print(df_name)
print("\n--- [2] 점수 데이터 (df_score) ---")
print(df_score)
print("\n--- [3] ID 기준 내부 조인 결과 ---")
print(df_merged)
```
