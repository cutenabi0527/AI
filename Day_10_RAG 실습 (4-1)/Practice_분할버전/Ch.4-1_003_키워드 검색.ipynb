```python
# TODO 0: 환경변수를 로드하고, 이전 실습에서 만든 all_documents를 준비해봅시다.

import os, glob
from dotenv import load_dotenv
from langchain_community.document_loaders import PyMuPDFLoader

load_dotenv()

# data/ 폴더의 모든 PDF 로드
all_documents = []
for pdf_path in sorted(glob.glob("data/*.pdf")):
    loader = PyMuPDFLoader(pdf_path)
    docs = loader.load()
    for doc in docs:
        doc.metadata["source_file"] = os.path.basename(pdf_path)
    all_documents.extend(docs)

print(f"PDF {len(glob.glob('data/*.pdf'))}개 → {len(all_documents)}페이지 로드 완료")
```

# 키워드 검색 (Keyword Search)
> 문서의 텍스트에서 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">정확히 일치하는 키워드</mark>를 찾아 관련 문서를 반환합니다.

> 키워드 검색은 문서에서 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">입력한 단어가 그대로 포함된 위치</mark>를 찾는 방식입니다.</br></br>
> 대표적인 알고리즘으로는 <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">TF-IDF(Term Frequency-Inverse Document Frequency)</mark>가 있는데, 단어가 특정 문서에서 얼마나 자주 등장하는지(TF)와 전체 문서에서 얼마나 희귀한지(IDF)를 곱하여 중요도를 산출합니다. </br></br>
> <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">BM25</mark>는 TF-IDF를 발전시킨 방식으로, 문서 길이를 보정하여 더 정교한 관련도 점수를 계산하며 현재도 많은 검색 엔진의 기반 알고리즘으로 쓰입니다.</br></br>
> 그러나 이 방식에는 명확한 한계가 있습니다. <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">동의어/유의어 처리 불가</mark>(예: "매출"과 "수익"은 같은 개념이지만 단어가 다르면 매칭되지 않음), <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">의미적 관계 파악 불가</mark>("회사 실적"처럼 직접 단어가 없어도 의미상 관련 있는 문서를 찾지 못함), <mark style="background-color:#FFF9C4; padding:2px 6px; border-radius:4px;">철자 변형 취약</mark>(띄어쓰기, 복수형, 활용형 차이에도 검색 결과가 달라짐) 등의 문제가 있습니다. 이번 실습에서는 이러한 원리와 한계를 직접 확인해봅니다.
